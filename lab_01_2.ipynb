```python
# Задание task_02_06.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

import math

a = float(input('Введите число для коэффициента a, не равняющееся нулю:'))
b = float(input('Введите число для коэффициента b:'))
c = float(input('Введите число для коэффициента c:'))

if a == 0:
    print("Ошибка: коэффициент a не может быть равен 0")
else:
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        x1 = (-b + discriminant**0.5) / (2*a)
        x2 = (-b - discriminant**0.5) / (2*a)
        print(f"x1 = {round(x1,1)}")
        print(f"x2 = {round(x2,1)}")
    elif discriminant == 0:
        x = -b / (2*a)
        print(f"x = {round(x,1)}")
    else:
        print("Решений нет")
```

```python
# Задание task_02_13.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

a = int(input("a = "))
b = int(input("b = "))

n_sum = 0
n_mult = 1
count = 0
odd_product = 1
odd_count = 0

for num in range(a, b + 1):
    n_sum += num
    n_mult *= num
    count += 1
    if num % 2 != 0:
        odd_product *= num
        odd_count += 1

n_avg = n_sum / count
if odd_count > 0:
    n_avg_geom = odd_product ** (1 / odd_count)
else:
    n_avg_geom = 0

print("Сумма =", n_sum)
print("Произведение =", n_mult)
print("Среднее арифметическое = {:.2f}".format(n_avg))
print("Среднее геометрическое нечетных чисел = {:.2f}".format(n_avg_geom))
```

```python
# Задание task_02_29.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

n = int(input("n = "))

nums = []
for i in range(1, n + 1):
    num = float(input("Введите {}-й элемент списка: ".format(i)))
    nums.append(num)

nums_pos = [x for x in nums if x > 0]
nums_neg = []
for x in nums:
    if x < 0:
        nums_neg.append(x)

sr_ar = sum(nums_pos) / len(nums_pos) if nums_pos else 0

if nums_neg:
    product = 1
    for num in nums_neg:
        product *= abs(num)
    sr_geom = product ** (1 / len(nums_neg))
else:
    sr_geom = 0

print("Исходный список:", nums)
print("Положительные числа:", nums_pos)
print("Отрицательные числа:", nums_neg)
print("Ср. арифм.: {:.2f}".format(sr_ar))
print("Ср. геом.: {:.2f}".format(sr_geom))
```

```python
# Задание task_02_30.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

# Все разделенные пробелом элементы будут преобразованы в список целых чисел
nums = [int(item) for item in input().split()]

# 1. Все положительные
for item in nums:
    if item <= 0:
        all_pos_1 = False
        break
else:
    all_pos_1 = True

all_pos_2 = all(item > 0 for item in nums)

# 2. Хотя бы 1 нулевой элемент
for item in nums:
    if item == 0:
        any_zero_1 = True
        break
else:
    any_zero_1 = False

any_zero_2 = any([item == 0 for item in nums])

# 3. Все четные
for item in nums:
    if item % 2 != 0:
        all_even_1 = False
        break
else:
    all_even_1 = True

all_even_2 = all(item % 2 == 0 for item in nums)

# 4. Хотя бы 1 нечетный элемент
for item in nums:
    if item % 2 != 0:
        any_odd_1 = True
        break
else:
    any_odd_1 = False

any_odd_2 = any([item % 2 != 0 for item in nums])

print("Все положительные:", all_pos_1, all_pos_2)
print("Хотя бы 1 нулевой элемент:", any_zero_1, any_zero_2)
print("Все четные:", all_even_1, all_even_2)
print("Хотя бы 1 нечетный элемент:", any_odd_1, any_odd_2)
```

```python
# Задание task_02_32.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

n = int(input("Количество рядов = "))

# 1. Заполнение мест
seats = []
for i in range(n):
    row = [int(item) for item in input().split()]
    seats.append(row)

# 2. Всего свободных мест
count = 0
for row in seats:
    for seat in row:
        if seat == 0:
            count += 1

print("Всего свободных мест:", count)

# 3. Ввод значений для поиска
# Учитывайте, что пользователь привык считать с 1, а не с 0,
# поэтому введенные сообщения необходимо скорректировать
n_p, m_p = [int(item) for item in
    input("Введите ряд и место через пробел: ").split()]

# n_p, m_p не должны выходить за границы
# Если выходит, необходимо указать, что "Такого места не существует"
if n_p < 1 or n_p > len(seats) or m_p < 1 or m_p > len(seats[n_p-1]):
    print("Такого места не существует")
else:
    # Корректируем индексы (пользователь вводит с 1, а в Python индексация с 0)
    is_free = (seats[n_p-1][m_p-1] == 0)
    print("Место свободно:", is_free)
```

```python
# Задание task_02_33.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

# 1. Заполнение списка
employees = [
    ["Петрова", "Анна", "Алексеевна", "Ж", 5],
    ["Семенов", "Николай", "Михайлович", "М", 2],
    ["Михайлов", "Игорь", "Васильевич", "М", 3]
]

# Ввод данных
n = int(input("Количество сотрудников = "))

employees = []
for i in range(n):
    data = input().split()
    employees.append([data[0], data[1], data[2], data[3], int(data[4])])

# Сортировка по стажу
sorted_by_exp = sorted(employees, key=lambda x: x[4])

print('Самый "молодой":', " ".join(sorted_by_exp[0][:3]))
print('Самый "старый":', " ".join(sorted_by_exp[-1][:3]))

# Разделение на мужчин и женщин
men = [emp for emp in employees if emp[3] == "М"]
women = [emp for emp in employees if emp[3] == "Ж"]

k = input("Введите букву начала имени = ").lower()

# Подсчет имен
men_count = sum(1 for emp in men if emp[1].lower().startswith(k))
women_count = sum(1 for emp in women if emp[1].lower().startswith(k))

# Сравнение
if men_count > women_count:
    print("У мужчин таких имен больше")
elif women_count > men_count:
    print("У женщин таких имен больше")
else:
    print("Количество таких имен одинаково")
```

```python
# Задание task_02_34.
#
# Выполнил: Миленин И.Д.
# Группа: АБП-231

# 1. Заполнение списка
#deposits = [
#    {"name": "Банк_1", "initial_sum": 50000, "rate": 5.2},
#    {"name": "Банк_2", "initial_sum": 20000, "rate": 3.7},
#    {"name": "Банк_3", "initial_sum": 45000, "rate": 6.8},
#]

n = int(input("Количество банков = "))

deposits = []

for i in range(n):
    data = input().split()
    deposit = {
        "name": data[0],
        "initial_sum": int(data[1]),
        "rate": float(data[2])
    }
    deposits.append(deposit)

# 2. Самый доступный банк (с наименьшей первоначальной суммой)
min_bank = min(deposits, key=lambda x: x["initial_sum"])

# 3. Самый выгодный банк
# прибыль = сумма * процент / 100
max_profit_bank = max(deposits, key=lambda x: x["initial_sum"] * x["rate"] / 100)
profit = max_profit_bank["initial_sum"] * max_profit_bank["rate"] / 100

print("Самый доступный банк: {}, первоначальная сумма: {:.2f} руб.".format(
    min_bank["name"], min_bank["initial_sum"]))
print("Самый выгодный банк: {}, прибыль: {:.2f} руб.".format(
    max_profit_bank["name"], profit))

# ---
# Пример вывода:
#
# Количество банков = 3
# Банк_1 50000 5.2
# Банк_2 20000 3.7
# Банк_3 45000 6.8
# Самый доступный банк: Банк_2, первоначальная сумма: 20000.00 руб.
# Самый выгодный банк: Банк_3, прибыль: 3060.00 руб.
```
